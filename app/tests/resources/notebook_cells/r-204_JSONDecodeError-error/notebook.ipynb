```r
# Load the package
library(jsonlite)
```

```r
# R Serialize to JSON string
# A named list (equivalent to a Python dictionary)
data <- list(
  task = "Write Python",
  number = 1,
  skills = c("python", "sql")
)

json_str <- toJSON(data, pretty = TRUE, auto_unbox = TRUE)

# Print the JSON string
print(json_str)
```

```r
# R Deserialize the JSON string
deserialized_data <- fromJSON(json_str)

# Print the deserialized data
print(deserialized_data)


original_data <- list(
  task = "Write Python",
  number = 1,
  skills = c("python", "sql")
)

stopifnot(identical(deserialized_data, original_data))
```
