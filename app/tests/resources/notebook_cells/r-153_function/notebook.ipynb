```r
library(dplyr)
library(readr)
```

```r
dataframe_file <- "dataframe.rds"
tibble_dataframe <- tibble::tibble(
    score = c(10, 20, 30),
    id = c(1, 2, 3)
  )
saveRDS(tibble_dataframe, dataframe_file)
```

```r
summary <- readRDS(dataframe_file) %>% summarise(across(everything(), sum, .names = "{.col}_total"))
print(summary)
```

```r
converted_to_characters_dataframe <- tibble_dataframe %>% mutate(across(score, base::as.character))
print(tibble_dataframe)
```
